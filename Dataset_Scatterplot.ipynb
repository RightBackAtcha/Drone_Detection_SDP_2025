```python
import numpy as np
import matplotlib.pyplot as plt

# Load the MFCC features and labels from .npy files
features = np.load(r"C:\Users\Jorda\Desktop\ECE 4830 Drone Detection SDP\dataset\mfcc_features.npy")  # Replace with your MFCC feature file
labels = np.load(r"C:\Users\Jorda\Desktop\ECE 4830 Drone Detection SDP\dataset\labels.npy")  # Replace with your labels file

# Check the shape of the data to understand the dimensions
print("Features Shape:", features.shape)
print("Labels Shape:", labels.shape)

# Example: Use the first two dimensions of the MFCC features for the scatter plot
# You can change which dimensions to plot by adjusting mfcc_features[:, 0] and mfcc_features[:, 1]
x = features[:, 0]  # First dimension
y = features[:, 1]  # Second dimension

# Create a scatter plot
plt.figure(figsize=(10, 6))

# Map unique labels (e.g., drone types) to different colors
unique_labels = np.unique(labels)
for label in unique_labels:
    # Select points that belong to this label
    indices = np.where(labels == label)
    plt.scatter(x[indices], y[indices], label=label, alpha=0.6)

# Add labels and title
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Features (First Two Dimensions)')

# Add a legend
plt.legend(title='Drone Labels')

# Show the plot
plt.show()
```
